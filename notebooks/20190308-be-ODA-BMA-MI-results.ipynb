```python
def get_analyte_names(df):
    column_names = df.columns
    return [get_analyte_name(int(column_name)) for column_name in column_names]
    
def get_analyte_name(analyte_id):
    analyte_name = str(analyte_id) + " "
    print(analyte_id)
    try:
        analyte_row = lab_value_types[ lab_value_types["ANALYSE"] == analyte_id].iloc[0]



        if analyte_row["BEZ"] is not None:
            analyte_name += str(analyte_row["BEZ"])

        if analyte_row["KBZ"] is not None and not pd.isna(analyte_row["KBZ"]):
            analyte_name += "(" + str(analyte_row["KBZ"]) + ")"

        if analyte_row["LOINC"] is not None:
            analyte_name += "=" + str(analyte_row["LOINC"])
    except:
        analyte_name = analyte_id
        
    return analyte_name
```

```python
for df in inc_prob_list:
    df.columns = get_analyte_names(df)
    df.sort_values(by=0, axis=1, ascending=False, inplace=True)
```

# Results

Search for 638 TNT=TNThsn, Creatin-Kinase=CK, 113 Lactat-Dehydrogenase=LDH to find the inclusion probabilities of the currently used predictors of Myocardial Infarction

```python
def plot_grouped_analytes_incprobs(dfs_list, fig_size=(15, 6)):
    incprobs_all = pd.concat(dfs_list, sort=False)
    incprobs_all.reset_index(inplace=True)
    incprobs_all = incprobs_all.transpose()
    incprobs_all.drop(["index"], inplace=True)

    ax = incprobs_all.plot.bar(figsize=fig_size)
```
